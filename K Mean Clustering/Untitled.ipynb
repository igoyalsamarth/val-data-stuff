```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from yellowbrick.cluster import SilhouetteVisualizer

df = pd.read_csv('dataset.csv')
X = np.array(df[['K','FKPer']])
#print(X)

model = KMeans(n_clusters = 2,init = 'k-means++',n_init = 50,max_iter = 1000,random_state = 42)
label = model.fit_predict(X)
#print(label)
print(model.inertia_)
u_label = np.unique(label)
#print(u_label)

for i in u_label:
    plt.scatter(X[label == i,0],X[label == i,1],label = i)


#filtered_label0 = X[label == 0]

#plt.scatter(filtered_label0[:,0],filtered_label0[:,1])
#sil_visualizer = SilhouetteVisualizer(model)

#plt.scatter(X[:,0],X[:,1])
#sil_visualizer.fit(X)
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], s = 250, marker = '*', c = 'r',label = 'Centroids')
```
