```python
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
from datetime import date
```

```python
xls = pd.ExcelFile('GODS-RNT.xlsx')

#-------------------INFO TO BE MANUALLY INPUTED----------------------

startRound = [1,13]
endRound = [12,18]
playerId = 'JoshS'
selfTeam = 'RNT'
opponentTeam = 'GODS'

#--------------------------------------------------------------------

df = pd.read_excel(xls,'Heatmaps')
df2 = pd.read_excel(xls,'Events')
maps = pd.read_excel(xls,'Matches Info')

matchMaps = np.array(maps['Maps']).reshape(-1)

locBreaks = [0]
locDataframe = {}

for i in range(len(df)-1):
    if(df.roundNumber[i] > df.roundNumber[i+1]):
        locBreaks.append(i+1)
locBreaks.append(len(df)+1)
#print(locBreaks)
for i in range(len(locBreaks)-1):
    locDataframe[i] = pd.DataFrame(df.iloc[locBreaks[i]:locBreaks[i+1]])
    #print(locDataframe[i])
    
eventsBreaks = [0]
eventsDataframe = {}

for i in range(len(df2)-1):
    if(df2.roundNumber[i] > df2.roundNumber[i+1]):
        eventsBreaks.append(i+1)
eventsBreaks.append(len(df2)+1)
#print(eventsBreaks)
for i in range(len(eventsBreaks)-1):
    eventsDataframe[i] = pd.DataFrame(df2.iloc[eventsBreaks[i]:eventsBreaks[i+1]])
    #print(eventsDataframe[i])
#print(len(eventsDataframe))
#------------------------MAIN CODE-----------------------------------
for i in range(len(locDataframe)):
    matchMap = matchMaps[i]
    defTeamId = []
    plotTime = []
    aR = []

    df = locDataframe[i][(locDataframe[i].playerId == playerId) & (locDataframe[i].roundNumber >= startRound[i]) & (locDataframe[i].roundNumber <= endRound[i])]
    df.reset_index(inplace = True)
    df = df[['roundNumber', 'playerId', 'roundTimeMillis', 'locationX','locationY']]

#-------------------Checking Coordinates--------------------------------------

    if(matchMap == 'Pearl'):
        siteB = np.array([[0,0],[565,0],[565,372],[572,372],[572,500],[545,500],[545,980],[0,980]])
        siteA = np.array([[567,0],[1000,0],[1000,980],[547,980],[547,502],[574,502],[574,370],[567,370]])
        siteC = np.array([[0,0]])
    elif(matchMap == 'Ascent'):
        siteB = np.array([[80,30],[595,30],[595,420],[573,420],[573,640],[537,640],[537,900],[80,900]])
        siteA = np.array([[597,30],[970,30],[970,900],[539,900],[539,642],[575,642],[575,422],[597,422]])
        siteC = np.array([[0,0]])
    elif(matchMap == 'Split'):
        siteB = np.array([[20,30],[485,30],[485,320],[550,320],[550,585],[497,585],[497,950],[20,950]])
        siteA = np.array([[487,30],[970,30],[970,950],[499,950],[499,587],[552,587],[552,318],[487,318]])
        siteC = np.array([[0,0]])
    elif(matchMap == 'Icebox'):
        siteB = np.array([[50,0],[510,0],[510,325],[484,325],[484,527],[458,527],[458,557],[465,557],[465,655],[560,655],[560,950],[50,950]])
        siteA = np.array([[512,0],[950,0],[950,950],[562,950],[562,653],[486,653],[486,554],[460,554],[460,529],[486,529],[486,327],[512,327]])
        siteC = np.array([[0,0]])
    elif(matchMap == 'Fracture'):
        siteB = np.array([[0,0],[495,0],[495,400],[485,400],[485,730],[550,730],[550,842],[440,842],[440,950],[0,950]])
        siteA = np.array([[497,0],[950,0],[950,950],[442,950],[442,844],[552,844],[552,728],[487,728],[487,402],[497,402],[497,0]])
        siteC = np.array([[0,0]])
    elif(matchMap == 'Bind'):
        siteB = np.array([[120,0],[520,0],[520,480],[500,480],[500,580],[550,580],[550,980],[120,980]])
        siteA = np.array([[522,0],[950,0],[950,980],[552,980],[552,578],[502,578],[502,482],[522,482]])
        siteC = np.array([[0,0]])
    elif(matchMap == 'Haven'):
        siteC = np.array([[50,100],[533,100],[533,258],[363,258],[363,402],[413,402],[413,583],[438,583],[438,713],[358,713],[358,810],[50,810]])
        siteB = np.array([[365,330],[400,330],[400,260],[653,260],[653,438],[578,438],[578,578],[573,578],[573,850],[580,850],[580,980],[360,980],[360,715],[440,715],[440,580],[415,580],[415,400],[365,400]])
        siteA = np.array([[535,30],[970,30],[970,800],[575,800],[575,580],[580,580],[580,440],[655,440],[655,258],[535,258]])
    elif(matchMap == 'Lotus'):
        siteC = np.array([[40,100],[550,100],[550,300],[418,300],[418,406],[338,406],[338,597],[428,597],[428,717],[448,717],[448,900],[40,900]])
        siteB = np.array([[420,408],[420,302],[655,302],[655,508],[590,508],[590,900],[450,900],[450,715],[430,715],[430,595],[340,595],[340,408]])
        siteA = np.array([[552,100],[970,100],[970,800],[592,800],[592,510],[657,510],[657,300],[552,300]])
#-------------------Calculating Anchor Time---------------------------------    
    
    xloc = np.array(df['locationX'])
    yloc = np.array(df['locationY'])
    r= np.unique(np.array(df['roundNumber']))
    time = np.array(df['roundTimeMillis'])
    #print(len(r))
    x = 0
    y = 0
    z = 0
    T = []
    for s in range(startRound[i],endRound[i]+1):
        #print(i)
        df1 = (df.loc[df['roundNumber'] == s])
        #print(df1)
        y = len(df1)
        if(y == 1):
            T.append(time[x]) #append1
            x = x+y
            continue
        if(((cv2.pointPolygonTest(siteB,(xloc[x],yloc[x]),False) != -1) == True) and (cv2.pointPolygonTest(siteB,(xloc[x+1],yloc[x+1]),False) != -1) == False):
            T.append(time[x]) #append2
            x = x+y
            continue
        if(((cv2.pointPolygonTest(siteA,(xloc[x],yloc[x]),False) != -1) == True) and (cv2.pointPolygonTest(siteA,(xloc[x+1],yloc[x+1]),False) != -1) == False):
            T.append(time[x]) #append3
            x = x+y
            continue
        if(((cv2.pointPolygonTest(siteC,(xloc[x],yloc[x]),False) != -1) == True) and (cv2.pointPolygonTest(siteC,(xloc[x+1],yloc[x+1]),False) != -1) == False):
            T.append(time[x]) #append4
            x = x+y
            continue
        for j in range(x,x+y):
            #print(j)
            if(((cv2.pointPolygonTest(siteB,(xloc[x],yloc[x]),False) != -1) == True) and ((cv2.pointPolygonTest(siteB,(xloc[j],yloc[j]),False) != -1) == False)):
                z = j-1
                break
            elif(((cv2.pointPolygonTest(siteA,(xloc[x],yloc[x]),False) != -1) == True) and ((cv2.pointPolygonTest(siteA,(xloc[j],yloc[j]),False) != -1) == False)):
                z = j-1
                break
            elif(((cv2.pointPolygonTest(siteC,(xloc[x],yloc[x]),False) != -1) == True) and ((cv2.pointPolygonTest(siteC,(xloc[j],yloc[j]),False) != -1) == False)):
                z = j-1
                break
            elif(((cv2.pointPolygonTest(siteA,(xloc[x],yloc[x]),False) != -1) == True) and ((cv2.pointPolygonTest(siteA,(xloc[x+y-1],yloc[x+y-1]),False) != -1) == True)):
                z = x+y-1
                break
            elif(((cv2.pointPolygonTest(siteB,(xloc[x],yloc[x]),False) != -1) == True) and ((cv2.pointPolygonTest(siteB,(xloc[x+y-1],yloc[x+y-1]),False) != -1) == True)):
                z = x+y-1
                break
            elif(((cv2.pointPolygonTest(siteC,(xloc[x],yloc[x]),False) != -1) == True) and ((cv2.pointPolygonTest(siteC,(xloc[x+y-1],yloc[x+y-1]),False) != -1) == True)):
                z = x+y-1
                break
        T.append(time[z])
        x = x+y
    
    T = np.array(T).reshape(-1).astype('float64')
    T = (T/(1000))
    #print(T.shape)
    avgT = round(np.average(T),2)
    #print(avgT)

#-------------------Calculating Round and Plant Time------------------------
    #print(i)
    roundTime = []
    plantTime = []
    df3 = eventsDataframe[i][(eventsDataframe[i].roundNumber >= startRound[i]) & (eventsDataframe[i].roundNumber <= endRound[i])]
    for x in range(startRound[i],endRound[i]+1):
        #print(i)
        df4 = (df3.loc[df3['roundNumber'] == x])
        roundTime.append(max(df4.roundTimeMillis))
        #print(df4)
        
        if('plant' in set(df4.eventType)):
            plantTime.append(df4.loc[df4['eventType'] == 'plant','roundTimeMillis'].iloc[0])
        else:
            plantTime.append(0)
    #print(len(roundTime))
    
    
    roundTime = np.array(roundTime).reshape(-1).astype('float64')
    plantTime = np.array(plantTime).reshape(-1).astype('float64')
    """
    df5 = df3[(df3.eventType == 'plant')]
    #df5 = df2[(df2.eventType == 'plant') & (df2.roundNumber >= startRound[i]) & (df2.roundNumber <= endRound[i])]
    
    for i in range(startRound[i],endRound[i]+1):
        print(i)
        if(i in set(df5.roundNumber)):
            plantTime.append(df5.loc[df5['roundNumber'] == i, 'roundTimeMillis'].iloc[0])
        else:
            plantTime.append(0)
    """
    for i in range(len(plantTime)):
        #print(i)
        if(plantTime[i] == 0):
            plantTime[i] = roundTime[i]
            
    for i in range(len(plantTime)):
        if T[i] > plantTime[i]:
            T[i] = plantTime[i]
        
    #print(plantTime)
    #print(len(plantTime))
    #plantTime = np.array(plantTime).reshape(-1).astype('float64')
    
#---------------------------Calculating Anchor Ratio-------------------------
    print(T)
    rT = ((roundTime/(1000)))
    pT = ((plantTime/1000))
    aR = []
    aR = T/pT
    for i in range(len(aR)):
        if aR[i] > 1:
            aR[i] = 1.0
    avgaR = round(np.average(aR),2)
    avgrT = round(np.average(rT),2)
    avgpT = round(np.average(pT),2)
    print(avgrT)

#--------------------------Printing Final Result-----------------------------
    today = date.today()
    d1 = today.strftime("%d/%m/%Y")
    print(f'The Anchor Time of {playerId} in {matchMap} is: {avgT}')
    print(f'The Anchor Ratio of {playerId} in {matchMap} is: {avgaR}')
    print(f'The Average Round Length of {selfTeam} vs {opponentTeam} in {matchMap}  with {selfTeam} on Defense is: {avgrT}')
    print(f'The Average Plant Time of {opponentTeam} in {matchMap} is: {avgpT}')
    
    anchorBuilder = [[d1, selfTeam, playerId, opponentTeam, matchMap, avgT, avgaR]]
    anchorBuilder = pd.DataFrame(anchorBuilder,columns=['date','selfTeam','playerName','oppoTeam','map','anchorTime','anchorRatio'])
    print(anchorBuilder)
    anchorBuilder.to_csv('Anchor Data.csv',index = False, mode = 'a',header = False)
    print("\n")
    durationBuilder = [[d1, selfTeam, opponentTeam, matchMap, avgrT, avgpT, (avgrT - avgpT)]] #calculation done for defense but printing results with attack for better understanding of the objective time.
    durationBuilder = pd.DataFrame(durationBuilder,columns=['date','selfTeam','oppoTeam','map','atkDuration','objDuration', 'retakeDuration'])
    print(durationBuilder)
    durationBuilder.to_csv('Duration Data.csv',index = False, mode = 'a',header = False)
    
```
