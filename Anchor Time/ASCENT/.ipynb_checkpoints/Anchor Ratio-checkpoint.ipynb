```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

newimage = cv2.imread('ascent.png')
oldimage = cv2.cvtColor(newimage,cv2.COLOR_BGR2RGB)
image = cv2.resize(oldimage,(1000,1000))
siteB = np.array([[80,30],[625,30],[625,420],[560,420],[560,900],[80,900]])
siteA = np.array([[630,100],[970,100],[970,900],[565,900],[565,420],[630,420]])
cv2.drawContours(image,[siteB], -1, (0,255,0), 2)
cv2.putText(image,text = 'Bomb Site B', org = (250,335), fontFace = cv2.FONT_HERSHEY_SIMPLEX,fontScale = 0.5, color = (0,255,0),thickness = 2,lineType = cv2.LINE_AA)
cv2.drawContours(image,[siteA], -1, (255,0,0), 2)
cv2.putText(image,text = 'Bomb Site A', org = (735,360), fontFace = cv2.FONT_HERSHEY_SIMPLEX,fontScale = 0.5, color = (255,0,0),thickness = 2,lineType = cv2.LINE_AA)
plt.imshow(image)
plt.axis('off')
plt.savefig('Ascent_Contour.png',dpi = 600,bbox_inches = 'tight',facecolor = 'auto')
#cv2.imshow('Contours',image)

#plt.imshow(image)
```

```python
df = pd.read_csv('C:\Drive D\DataScience\Anchor Time\ASCENT\GEvsGRE\hoax_GEvsGRE_Ascent_loc.csv')
xloc = np.array(df['locationX'])
yloc = np.array(df['locationY'])
r= np.array(df['roundNumber'])
time = np.array(df['roundTimeMillis'])
#print(len(r))
x = 0
y = 0
z = 0
T = []
for i in range(13,21): #change 13 to (whatever the final number of rounds are in second half - 12)
    #print(i)
    df1 = (df.loc[df['roundNumber'] == i]) # == (i+12) for second half
    #print(df1)
    y = len(df1)
    #print('x',x)
    #print('x+y',x+y-1)
    if(y == 1):
        T.append(time[x]) #append1
        x = x+y
        continue
    if(((cv2.pointPolygonTest(siteB,(xloc[x],yloc[x]),False) != -1) == True) and (cv2.pointPolygonTest(siteB,(xloc[x+1],yloc[x+1]),False) != -1) == False):
        T.append(time[x]) #append2
        x = x+y
        continue
    if(((cv2.pointPolygonTest(siteA,(xloc[x],yloc[x]),False) != -1) == True) and (cv2.pointPolygonTest(siteA,(xloc[x+1],yloc[x+1]),False) != -1) == False):
        T.append(time[x]) #append3
        x = x+y
        continue
    for j in range(x,x+y):
        #print(j)
        if(((cv2.pointPolygonTest(siteB,(xloc[x],yloc[x]),False) != -1) == True) and ((cv2.pointPolygonTest(siteB,(xloc[j],yloc[j]),False) != -1) == False)):
            z = j-1
            break
        elif(((cv2.pointPolygonTest(siteA,(xloc[x],yloc[x]),False) != -1) == True) and ((cv2.pointPolygonTest(siteA,(xloc[j],yloc[j]),False) != -1) == False)):
            z = j-1
            break    
        elif(((cv2.pointPolygonTest(siteA,(xloc[x],yloc[x]),False) != -1) == True) and ((cv2.pointPolygonTest(siteA,(xloc[x+y-1],yloc[x+y-1]),False) != -1) == True)):
            z = x+y-1
            break
        elif(((cv2.pointPolygonTest(siteB,(xloc[x],yloc[x]),False) != -1) == True) and ((cv2.pointPolygonTest(siteB,(xloc[x+y-1],yloc[x+y-1]),False) != -1) == True)):
            z = x+y-1
            break
    #print('z = ',z)
    T.append(time[z])
        
    x = x+y
T = np.array(T)
T = ((T/(1000)))
print(T.shape)
anchor = round(np.average(T),2)
print(anchor)
y = np.arange(13,21,1) #arange(13,25,1) for second half
fig = plt.figure(figsize = (18,9),facecolor = '#fff3d1')
ax = plt.axes()
ax.set_facecolor('#fff3d1')
plt.grid(color = '#a3a3a3', linestyle='--',linewidth=0.2)
#plt.ylim([0,120])
plt.plot(y,T,label = 'Round Anchor Time')
plt.hlines(anchor,xmin = 13,xmax = 20,color = 'red',linestyle = ':',label = 'Average Anchor Time') #xmin = 13,xmax = 24
plt.annotate(f'Avg. Anchor Time = {anchor}s', (18.8,anchor+1),size = 14)
plt.title('Gods Hoax || Site Anchor Time || ASCENT || vs GE',size = 18)
plt.xlabel('Round number ->',size = 16)
plt.ylabel('Time [s] ->',size = 16)

plt.legend(fontsize = '12',loc = 'lower right')
plt.savefig('hoax_GEvsGRE_Ascent_AT.png',dpi = 600,bbox_inches = 'tight',facecolor = 'auto')
plt.show()
```

```python
df2 = pd.read_csv('C:\Drive D\DataScience\Anchor Time\ASCENT\GEvsGRE\Ember_GEvsGRE_Ascent_rounds.csv')
roundTime = np.array(df2['roundDurationMillis'])
plantTime = np.array(df2['spikePlantRoundTimeMillis'])
cond = np.array(df2['spikePlant'])
#print(roundTime)
rT = ((roundTime/(1000)))
pT = ((plantTime/1000))
aR = []
for i in range(len(df2)):
    if cond[i] == True:
        aR.append(T[i]/pT[i])
    else:
        aR.append(T[i]/rT[i])

print(T)
print(pT)
print(aR)
#print(T/rT)
for i in range(len(aR)):
    if aR[i] > 1:
        aR[i] = 1.0
#print(aR)
y = np.arange(13,21,1)
fig = plt.figure(figsize = (18,9),facecolor = '#fff3d1')
ax = plt.axes()
ax.set_facecolor('#fff3d1')
plt.grid(color = '#a3a3a3', linestyle='--',linewidth=0.2)
plt.plot(y,aR,label = 'Round Anchor Ratio')
avgaR = round(np.average(aR),2)
print(avgaR)
plt.hlines(avgaR,xmin = 13,xmax = 20,color = 'red',linestyle = ':',label = 'Average Anchor Ratio')
plt.annotate(f'Avg. Anchor Ratio = {avgaR}s', (17.4,avgaR+0.01),size = 14)
plt.title('Gods Hoax || Site Anchor Ratio || ASCENT || vs GE',size = 18)
plt.xlabel('Round number ->',size = 16)
plt.ylabel('Anchor Ratio ->',size = 16)
plt.legend(fontsize = '12',loc = 'lower right')
plt.savefig('hoax_GEvsGRE_Ascent_AR.png',dpi = 600,bbox_inches = 'tight',facecolor = 'auto')

#print(anchor/avgrT)
```
